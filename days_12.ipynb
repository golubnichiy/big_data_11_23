```python
def convert1(obj):
    count = 0
    L = []
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i["name"])
            count += 1
        else:
            break
    return L
```

```python
df_credits["cast"] = df_credits["cast"].apply(convert1)
```

```python
df_credits["crew"][1]
```

```python
def convert_dir(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
        else:
            continue
    return L
```

```python
df_credits["director"] = df_credits["crew"].apply(convert_dir)
```

```python
df1 = fin_df[["id", "title", "overview", "genres", "keywords", "cast", "director"]]
df1.head(5)
```

```python
df1["genres"] = df1["genres"].apply(lambda x:[i.replace(" ", "") for i in x])
df1["keywords"] = df1["keywords"].apply(lambda x:[i.replace(" ", "") for i in x])
df1["cast"] = df1["cast"].apply(lambda x:[i.replace(" ", "") for i in x])
df1["director"] = df1["director"].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
df1.head()
```

```python
df1["strings"] = df1["overview"]+df1["genres"]+df1["keywords"]+df1["cast"]+df1["director"]
```

```python
df1.head()
```

```python
new_df = df1[["id", "title", "strings"]]
new_df.head()
```

```python
new_df["strings"] = new_df["strings"].apply(lambda x: " ".join(x))
new_df["strings"] = new_df["strings"].apply(lambda x: x.lower())
new_df.head()
```

### Векторизация

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["strings"]).toarray()
```

```python
vectors.shape
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["strings"] = new_df["strings"].apply(stem)
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
vectors = cv.fit_transform(new_df["strings"]).toarray()
ps=PorterStemmer()
new_df["strings"] = new_df["strings"].apply(stem)
```

```python
new_df["strings"][31]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Ivan Tsarevich & the Grey Wolf 2")
```
