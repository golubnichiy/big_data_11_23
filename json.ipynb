### Импортирование библиотек

```python
import pandas as pd
import json
```

```python
with open("data.json", "r") as f:
    data = json.load(f)
```

### Создание dataframe

```python
df = pd.json_normalize(data)
df.head(3)
```

```python
df["created"] = pd.to_datetime(df["created"])
```

```python
print(df)
```

```python
df.describe()
```

```python
df_items = df.explode("items")
df_items
```

```python
total_sales_by_unit = df_items.groupby("businessUnit")['items'].apply(lambda x: sum(item["quantity"] * item["price"] for item in x)).reset_index(name="total_sales")
total_sales_by_unit
```

```python
sales_by_date = df_items.groupby(df["created"].dt.date)['items'].apply(lambda x: sum(item["quantity"] * item["price"] for item in x)).reset_index(name="daily_sales")
sales_by_date
```
