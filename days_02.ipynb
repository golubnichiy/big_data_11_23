### Визуализация данных

```python
import pandas as pd
data = pd.read_csv("HWS.csv")
print(data)
```

#### Построение гистрограммы

```python
import matplotlib.pyplot as plt
data["Height"].hist()
plt.show()
```

```python
data["Height"].hist(bins=8)
plt.show()
```

```python
data["Height"].hist(bins=40)
plt.show()
```

### Столбчатый график (barplot)

```python
avg_height_by_index = data.groupby("Index")["Height"].mean()
print(avg_height_by_index)
```

```python
avg_height_by_index.plot(kind="bar")
plt.show()
```

```python
avg_height_by_index.plot(kind="bar", title="Средний рост по группам индекса массы тела")
plt.show()
```

```python
aq = pd.read_csv("aq.csv")
print(aq)
```

```python
aq.plot(x="Date", y="NO", kind="line")
plt.show()
```

```python
aq.plot(x="Date", y="NO", kind="line", rot=30)
plt.show()
```

```python
data.plot(x="Height", y="Weight", kind="scatter")
plt.show()
```

```python
data[data["Gender"]=="Male"]["Height"].hist(alpha=0.6)
data[data["Gender"]=="Female"]["Height"].hist(alpha=0.6)
plt.legend(["Male", "Female"])
plt.show()
```

### Создание датафреймов

- Список словарей
- Словарь списков

```python
list_of_dicts = [
{"Name":"Сноу", "Breed":"Лабрадор", "Height":35},
{"Name":"Макс", "Breed":"Бульдог", "Height":46}
]
data2 = pd.DataFrame(list_of_dicts)
print(data2)
```

```python
dict_of_lists = {
"Name":["Сноу","Макс"], "Breed":["Лабрадор", "Бульдог"], "Height":[35,46]}
data3 = pd.DataFrame(dict_of_lists)
print(data3)
```
