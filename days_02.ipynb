### Визуализация данных

```python
import pandas as pd
data = pd.read_csv("HWS.csv")
print(data)
```

#### Построение гистрограммы

```python
import matplotlib.pyplot as plt
data["Height"].hist()
plt.show()
```

```python
data["Height"].hist(bins=8)
plt.show()
```

```python
data["Height"].hist(bins=40)
plt.show()
```

### Столбчатый график (barplot)

```python
avg_height_by_index = data.groupby("Index")["Height"].mean()
print(avg_height_by_index)
```

```python
avg_height_by_index.plot(kind="bar")
plt.show()
```

```python
avg_height_by_index.plot(kind="bar", title="Средний рост по группам индекса массы тела")
plt.show()
```
